```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 1000)
import scipy as sp
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

data = pd.read_csv("Churn.csv")
print(data['churn'].unique())
print(data.dtypes)    ## it is important to note the dtype of column churn as it is
                          # type bool and not object, so we have to be careful when we are using np.where
data['churn'] = np.where(data['churn'] == False, 0, 1)
print(data.dtypes)
# print(data['churn'].head(100))
print(data.head(5))
print(data.columns)
print(data.describe())
print('The number of rows and columns in the data set are\n')
print(data.shape)
print(data['total day minutes'].mean())
groupchurned = data.groupby('churn')
print(groupchurned['churn'].count())
print(groupchurned.count())
print(groupchurned['total day minutes'].count())
print(groupchurned['total day minutes'].mean())
print(data.describe())
print(data.max())
print(data.nunique()) #To find the number of unique values in the data set
m = data.mean()
print(m)
data.fillna(data.mean(), inplace=True)
print(data.head(10))

print(data.nunique())
col1 = ['international plan', 'voice mail plan']  #Cleaning the features binary variable
col2 = ['number vmail messages', 'total day minutes', 'total day calls', 'total day charge', 'total eve minutes',
        'total eve calls', 'total eve charge', 'total night minutes', 'total night calls', 'total night charge',
        'total intl calls', 'total intl minutes', 'total intl charge', 'customer service calls']
data1 = data[col1]
data1 = pd.get_dummies(data1)
data2 = data[col2]
data1.fillna(data1.min(), inplace=True)
data2.fillna(data2.mean(), inplace=True)

from sklearn.preprocessing import StandardScaler
transformer = StandardScaler().fit(data2)
data2_standardized = pd.DataFrame(transformer.transform(data2))
# print(x_standardized.shape)

X = pd.concat([data1, data2_standardized], axis=1) #Getting the features together
Y = data['churn']
X_train, X_test = train_test_split(X, test_size=0.3)
Y_train, Y_test = train_test_split(Y, test_size=0.3)

print(X_train.head())

for i in range(1, 50):
    knn = KNeighborsClassifier(n_neighbors=i, p=2)
    knn.fit(X_train, Y_train)
    y_pred = knn.predict(X_test)
    print('The accuracy value for neighbors', i)
    print(accuracy_score(Y_test, y_pred))
```
